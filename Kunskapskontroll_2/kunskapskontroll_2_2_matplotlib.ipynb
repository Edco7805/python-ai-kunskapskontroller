```python
# Matplotlib kan användas två olika gränssnitt: ett explicit, objetorienterat och ett implicit, 
# funktionellt. 
 

# 1. Pyplot-gränssnittet (state-based) – använder plt-kommandon som plt.plot(), 
# plt.xlabel(), osv. Det håller automatiskt reda på den aktiva figuren och axlarna.

# 2. Objektorienterat gränssnitt (explicit Axes-baserat) – skapar explicit figurer 
# och axlar med fig, ax = plt.subplots() och använder sedan metoder på ax som ax.plot(),
# ax.set_xlabel(), osv.

# Rekommenderat tillvägagångssätt är det objektorienterade gränssnittet, eftersom det ger 
# tydligare kontroll, är lättare att hantera i mer komplexa diagram, och fungerar 
# bättre i skript och större applikationer.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Skapa data
x = np.linspace(-4, 4, 100)  # 100 punkter mellan -4 och 4
y = x**2

# Skapa figur och axlar
fig, ax = plt.subplots(figsize=(8, 6))

# Rita grafen
ax.plot(x, y, 'b-', linewidth=2, label='y = x²')

# Lägg till titel och etiketter
ax.set_title('Parabel: y = x²', fontsize=14, fontweight='bold')
ax.set_xlabel('x-värden', fontsize=12)
ax.set_ylabel('y = x²', fontsize=12)

# Lägg till rutnät och legend
ax.grid(True, alpha=0.3)
ax.legend()
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)

# Visa diagrammet
plt.tight_layout()
plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Skapa data
x = np.linspace(-4, 4, 100)
y = x**2

# Skapa en ny figur
plt.figure(figsize=(8, 6))

# Rita grafen
plt.plot(x, y, 'r-', linewidth=2, label='y = x²')

# Lägg till titel och etiketter
plt.title('Parabel: y = x²', fontsize=14, fontweight='bold')
plt.xlabel('x-värden', fontsize=12)
plt.ylabel('y = x²', fontsize=12)

# Lägg till rutnät och legend
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Anpassa layout och visa
plt.tight_layout()
plt.show()
```
