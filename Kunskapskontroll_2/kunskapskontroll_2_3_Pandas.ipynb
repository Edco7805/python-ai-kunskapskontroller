### Calculate the mean of each numeric column. 

```python
import pandas as pd
df = pd.read_csv("cars_data.csv")

means = df.mean(numeric_only=True)
means_df = pd.DataFrame({'Kolumn': means.index, 'Medelvärde': means.values})
print(means_df)
```

### Select the rows where the column "company" is equal to 'honda'. 

```python
df = pd.read_csv("cars_data.csv")
honda_company = df [df ["company"] == "honda"]
print(honda_company)
```

### Sort the data set by price in descending order. This should *not* be an inplace operation. 

```python

import pandas as pd
df = pd.read_csv("cars_data.csv")
sorted_df = df.sort_values(by='price', ascending=False)
print(sorted_df)
print("\n--- Original DataFrame (oförändrad) ---")
print(df.head())

```

### Select the rows where the column "company" is equal to any of the values (audi, bmw, porsche).

```python
import pandas as pd
df = pd.read_csv("cars_data.csv")
# Skapa en lista med önskade märken
desired_companies = ["audi", "bmw", "porsche"]  
filtered_df = df[df["company"].isin(desired_companies)]
print(filtered_df)
```

### Find the number of cars (rows) for each company. 

```python
import pandas as pd
df = pd.read_csv("cars_data.csv")

cars_per_company = df["company"].value_counts()
result_df = pd.DataFrame({
    'Företag': cars_per_company.index,
    'Antal bilar': cars_per_company.values
})

print("Number of cars for each company:")
print(result_df.to_string(index=False))  
print(f"\nTotalt antal bilar i datasetet: {len(df)}")
```

### Find the maximum price for each company. 

```python
import pandas as pd
df = pd.read_csv("cars_data.csv")
max_price_per_company = df.groupby('company')['price'].max()
print("Maximum price for each company:")
print(max_price_per_company)
```
